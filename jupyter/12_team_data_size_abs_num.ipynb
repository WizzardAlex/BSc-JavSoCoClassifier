# Overall performance for number of files

```python
train_size = ['5','10','15','20','25']
y_axis = {
    'naive': naive_accuracy_test2,
    'maxEnt': maxEnt_accuracy_test2,
    'svm': svm_accuracy_test2
}
#########ACCURACY#############################

Graphs.plot_line_graph("Accuracy over abs.train size/concern", train_size, y_axis, "Train size/concern", "Accuracy", type ="accuracy")
##############################################

##############PRECISION#######################
Graphs.plot_line_graph("Macro avg. precision over abs.train size/concern", train_size, {'naive': naive_precision_test2,
                                                             'maxEnt': maxEnt_precision_test2,
                                                             'svm': svm_precision_test2
                                                            }, "Train size/concern", "Macro.avg. precision", type = "macro")

Graphs.plot_line_graph("Weighted avg. precision over abs.train size/concern", train_size, {'naive': naive_precision_test2,
                                                             'maxEnt': maxEnt_precision_test2,
                                                             'svm': svm_precision_test2
                                                            }, "Train size/concern", "Weighted.avg. precision", type = "weighted")

##############RECALL#######################
Graphs.plot_line_graph("Macro avg. recall over abs.train size/concern", train_size, {'naive': naive_recall_test2,
                                                             'maxEnt': maxEnt_recall_test2,
                                                             'svm': svm_recall_test2
                                                            }, "Train size/concern", "Macro.avg. recall", type = "macro")

Graphs.plot_line_graph("Weighted avg. recall over abs.train size/concern", train_size, {'naive': naive_recall_test2,
                                                             'maxEnt': maxEnt_recall_test2,
                                                             'svm': svm_recall_test2
                                                            }, "Train size/concern", "Weighted.avg. recall", type = "weighted")


Graphs.plot_line_graph("Macro avg. F1-score over abs.train size/concern", train_size, {'naive': naive_f1_test2,
                                                             'maxEnt': maxEnt_f1_test2,
                                                             'svm': svm_f1_test2
                                                            }, "Train size/concern", "Macro.avg. F1-score", type = "macro")

Graphs.plot_line_graph("Weighted avg. F1-score over abs.train size/concern", train_size, {'naive': naive_f1_test2,
                                                             'maxEnt': maxEnt_f1_test2,
                                                             'svm': svm_f1_test2
                                                            }, "Train size/concern", "Weighted.avg. F1-score", type = "weighted")

```

```python
size_type='abs'

maxEnt_accuracy_test = maxEnt_accuracy_test2
svm_accuracy_test = svm_accuracy_test2
naive_accuracy_test =  naive_accuracy_test2

maxEnt_precision_test = maxEnt_precision_test2
svm_precision_test = svm_precision_test2
naive_precision_test = naive_precision_test2

maxEnt_recall_test = maxEnt_recall_test2
svm_recall_test = svm_recall_test2
naive_recall_test = naive_recall_test2

maxEnt_f1_test = maxEnt_f1_test2
svm_f1_test = svm_f1_test2
naive_f1_test = naive_f1_test2

df_maxEnt = pd.DataFrame({
    'accuracy': maxEnt_accuracy_test, 
    'precision' : [x[1] for x in maxEnt_precision_test],
    'recall': [x[1] for x in maxEnt_recall_test],
    'f1' : [x[1] for x in maxEnt_f1_test],
    'size' : train_size
    })

df_maxEnt['classifier']='maxEnt'

df_svm = pd.DataFrame({
    'accuracy': svm_accuracy_test, 
    'precision' : [x[1] for x in svm_precision_test],
    'recall': [x[1] for x in svm_recall_test],
    'f1' : [x[1] for x in svm_f1_test],
    'size' : train_size
    })
df_svm['classifier']='svm'

df_naive = pd.DataFrame({
    'accuracy': naive_accuracy_test, 
    'precision' : [x[1] for x in naive_precision_test],
    'recall': [x[1] for x in naive_recall_test],
    'f1' : [x[1] for x in naive_f1_test],
    'size' : train_size
    })
df_naive['classifier']='naive'

df = pd.concat([df_maxEnt, df_naive, df_svm])
df['system']=system
df['size type']=size_type

fn = 'size_' + size_type + '_' + system + '.csv'
fp = Path.cwd().parent/'Data'/system/fn

df.to_csv(fp, index=False)
```
