```python
# Imports and scripts
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import plot_confusion_matrix
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import pathmagic
%matplotlib inline
with pathmagic.context():
    import Preprocess as Prep
    import RelativePaths as RP
    import Evaluation as Eva
    import GatherData as Gather
    import Graphs
    import Utils

system = 'team'
```

```python
# File containing settings
path_to_yaml = '../config.yaml'
config = Utils.read_yaml_file(path_to_yaml)

files = config['file locations'][system]
preprocess_settings = config['preprocess settings list']
```

```python
# set file_locations which will be relative to computer in use
from pathlib import Path

system_folder = str(Path.cwd().parent / files['system folder'])

raw_data_csv = str(Path.cwd().parent / files['raw data'])
tmp_csv = str(Path.cwd().parent / files['tmp data'])

processed_data_csv = str(Path.cwd().parent / files['data size percentage'])
```

```python
Gather.gather_architectural_concerns_data(system_folder, raw_data_csv)
```

```python
dataset_df = pd.read_csv(raw_data_csv)
y_labels = dataset_df.Label.unique()
x_quantity = [len(dataset_df.loc[dataset_df['Label']==label]) for label in y_labels]
tmp_df = pd.DataFrame({
    'Labels' : y_labels,
    'Quantity' : x_quantity
})
tmp_df = tmp_df.sort_values(by=['Quantity'])
plt.style.use("seaborn-whitegrid")
plt.barh(y=tmp_df.Labels, width=tmp_df.Quantity)
for i, v in enumerate(tmp_df.Quantity):
    plt.text(v, i, str(v), color='black', fontweight='bold', ha='left', va='center')

plt.xlabel('Quantities')
plt.ylabel('Labels')
plt.title('Shows the number of files for every label, from JabRef')
```

# Overall performance for test_size

```python
train_size = ['10%','15%','20%','25%']
y_axis_accuracy = {
    'naive': naive_accuracy_test1,
    'maxEnt': maxEnt_accuracy_test1,
    'svm': svm_accuracy_test1
}
#########ACCURACY#############################

Graphs.plot_line_graph("Accuracy over train size (%)", train_size, y_axis_accuracy, "Train size (%)", "Accuracy", type ="accuracy")
##############################################

##############PRECISION#######################
Graphs.plot_line_graph("Macro avg. precision over train size (%)", train_size, {'naive': naive_precision_test1,
                                                             'maxEnt': maxEnt_precision_test1,
                                                             'svm': svm_precision_test1
                                                            }, "Train size (%)", "Macro.avg. precision", type = "macro")

Graphs.plot_line_graph("Weighted avg. precision over train size (%)", train_size, {'naive': naive_precision_test1,
                                                             'maxEnt': maxEnt_precision_test1,
                                                             'svm': svm_precision_test1
                                                            }, "Train size (%)", "Weighted.avg. precision", type = "weighted")

##############RECALL#######################
Graphs.plot_line_graph("Macro avg. recall over train size (%)", train_size, {'naive': naive_recall_test1,
                                                             'maxEnt': maxEnt_recall_test1,
                                                             'svm': svm_recall_test1
                                                            }, "Train size (%)", "Macro.avg. recall", type = "macro")

Graphs.plot_line_graph("Weighted avg. recall over train size (%)", train_size, {'naive': naive_recall_test1,
                                                             'maxEnt': maxEnt_recall_test1,
                                                             'svm': svm_recall_test1
                                                            }, "Train size (%)", "Weighted.avg. recall", type = "weighted")

Graphs.plot_line_graph("Macro avg. F1-score over train size (%)", train_size, {'naive': naive_f1_test1,
                                                             'maxEnt': maxEnt_f1_test1,
                                                             'svm': svm_f1_test1
                                                            }, "Train size (%)", "Macro.avg. F1-score", type = "macro")

Graphs.plot_line_graph("Weighted avg. F1-score over train size (%)", train_size, {'naive': naive_f1_test1,
                                                             'maxEnt': maxEnt_f1_test1,
                                                             'svm': svm_f1_test1
                                                            }, "Train size (%)", "Weighted.avg. F1-score", type = "weighted")
```

```python
size_type='ratio'

maxEnt_accuracy_test1
svm_accuracy_test1
naive_accuracy_test1

maxEnt_precision_test1
svm_precision_test1
naive_precision_test1

maxEnt_recall_test1
svm_recall_test1
naive_recall_test1

maxEnt_f1_test1
svm_f1_test1
naive_f1_test1

df_maxEnt = pd.DataFrame({
    'accuracy': maxEnt_accuracy_test1, 
    'precision' : [x[1] for x in maxEnt_precision_test1],
    'recall': [x[1] for x in maxEnt_recall_test1],
    'f1' : [x[1] for x in maxEnt_f1_test1],
    'size' : train_size
    })

df_maxEnt['classifier']='maxEnt'

df_svm = pd.DataFrame({
    'accuracy': svm_accuracy_test1, 
    'precision' : [x[1] for x in svm_precision_test1],
    'recall': [x[1] for x in svm_recall_test1],
    'f1' : [x[1] for x in svm_f1_test1],
    'size' : train_size
    })
df_svm['classifier']='svm'

df_naive = pd.DataFrame({
    'accuracy': naive_accuracy_test1, 
    'precision' : [x[1] for x in naive_precision_test1],
    'recall': [x[1] for x in naive_recall_test1],
    'f1' : [x[1] for x in naive_f1_test1],
    'size' : train_size
    })
df_naive['classifier']='naive'

df = pd.concat([df_maxEnt, df_naive, df_svm])
df['system']=system
df['size type']=size_type

fn = 'size_' + size_type + '_' + system + '.csv'
fp = Path.cwd().parent/'Data'/system/fn

df.to_csv(fp, index=False)
```
